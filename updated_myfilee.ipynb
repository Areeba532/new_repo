```python
def generate_word_doc(name,Email,intro,exp,educ,skill):
    # Create a new Word document
    doc = Document()

    """new_parser = HtmlToDocx()

    html = '<b><h1>CURRICULUM VITAE</h1></b><hr>'
    html='<ul><li></li></ul>'
    new_parser.add_html_to_document(html, doc)"""

    run = doc.add_paragraph().add_run()
    style = doc.styles['Normal']
    font = style.font
    font.name = 'Times New Roman'
    font.size = Pt(12)
    font.color.rgb = RGBColor(0,0,0)
    
    # HEADING
    head=doc.add_heading('CURRICLUM VITAE', 0)
    run = head.runs[0]
    head.runs[0].font.bold = True
    run.font.size = Pt(24)
    run.text = run.text.upper()
    font.name = 'Georgia'
    run.font.color.rgb = RGBColor(0,128,128)

    # NAME
    paragraph=doc.add_paragraph(f"{name}")
    run = paragraph.runs[0]
    paragraph.runs[0].font.bold = True
    run.font.size = Pt(16)
    run.text = run.text.upper()
    font.name = 'Georgia'

     # EMAIL
    email=doc.add_paragraph(f"{Email}")
    email.runs[0].font.size = Pt(12)

    # OBJECTIVE
    obj=doc.add_heading('OBJECTIVE',level=1)
    personal=conver(f" write a short 50 words objective for the job i am applying {intro}. it must looks professional ")
    doc.add_paragraph(personal)
    obj.runs[0].font.bold = True



    #EXPERIENCE
    exp=doc.add_heading('WORK EXPERIENCE', level=1)
    exper=conver(f" write the 3 most important experience for {intro} cv. write each point in a dot bullets ")
    doc.add_paragraph(exper)


    #EDUCATION
    doc.add_heading('EDUCATION', level=1)
    doc.add_paragraph(educ)

    # SKILLS
    doc.add_heading('SKILLS', level=1)
    skills_2=conver(f" write the name of 3 most important skills  for {intro} and must be related with {skill}. it must looks professional and name each skill in dot bullets")
    doc.add_paragraph(skills_2)
    
    # Save the document
    doc_file = "output_document.docx"
    doc.save(doc_file)
    
    return doc_file
```
